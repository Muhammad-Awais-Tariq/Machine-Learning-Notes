## K means clustering

The most common type of clustering algorithm that is used today is k mean clustering

Example

![](Images/clustering_example.png)

If we have data like this and we want to devide it into two clusters the algorithm will work as

![](Images/first_step.png)

In the first step the algorithm take the intial guess of where the center of the two clusters are and its not an accurate guess as it does not have to be

K means will do two things always:
1. Assign different points to cluster centeroids (these cross which it think is cluster center)
2. Move the cluster centeroids to the center of data

The very first thing the k mean is doing that its guessing the center of the data called the cluster centroids

![](Images/centeroids.png)

After the algorithm has taken the intial guess at where it think the center is it will go through all the data points and for each point it will check to which cluster centriod that point is closer to and it will assign the point to that cluster

![](Images/second_step.png)

At the second step the algorithm will look at all the red points and take the average of all of those red points or a group of points and move the cluster cenetorid at that point 

![](Images/finding_avearge.png)

Now it will do the similar thing for the blue dots as well}

![](Images/reclculating_cluster_center.png)

Now onces it done and we have better guess for the center of the data we will look again at all the data and repeat the steps assigning the points that is closet to one center averaging and moving the centeroid again

![](Images/recolored.png)

Onces based on the ceneteroid the points are recolored we will again find the average to move the centeroid there


![](Images/again_moving_center.png)

Now we will  reapeat again first check the average location of all the points and move the ceneteroid to the average

![](Images/new_location.png)

And if keep on repeating these points and finally a point will reach where there is no change in the color of the points or the position of the cluster ceneteroids this mean that at this step k means algorithm has converged and we have done our function
