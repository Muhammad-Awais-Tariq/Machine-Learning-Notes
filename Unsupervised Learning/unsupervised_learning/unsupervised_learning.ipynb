## Cost function

In case of the k mean clustering the cost is the average of the distance of the training example to the cenetroid of the cluster for that example squared

![](Images/cost_function.png)

The k mean algorithm is trying to find the location of the centeroid as well as the assignment of the point to the cenetroid that minimize the distance between the both

![](Images/example_cost.png)

In the above example the cost function will look from the cenetroid and start calculating all the distance between these examples squared and then find the avearage

Distortation is the cost function function

In the first step of the k mean algorithm when we assign the points to the centeroids we are trying to minimize the cost as much as possible by updating the values of the points that mean the c

In the second step when we move the cluster we uodate the mue to minimize the cost function

![](Images/example_step_1.png)

If we have a simple example like two centeroids and a single point what the algorithm does in the first step is the assign it to cenetroid with the lowest distance meaning the lowest cost which in this case will be red one

![](Images/example_second_step.png)

In the second step as we can see by the red that if our cenetroid is not at the center then we will have a large cost function but if we move it to the center or in other word mean of the cluster then the overall cost function will be reduction

The cost function or distortion for the k mean algorithm should always go down or in the worst case stay same if it ever goes up there is a bug in the code

If the function stops going down this mean that the k means has converged and we should stop
