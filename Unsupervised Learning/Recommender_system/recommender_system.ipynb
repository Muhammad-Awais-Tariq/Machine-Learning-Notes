## Mean normalization

Just like the linear algorithm if we normalize the feature before applying the model it would improve its performance same is the case with the recommender system if we normalize the rating before we run the algorithm it will also increase the performance of algorithm drastically

![](Images/without_normalization.png)

If we have a fifth user who hvnt rated any movie and we apply the cost function without the normalization the function will predict the parameter for it to be 0 0

Based on these the algorithm will predict that he will give the rating of 0 to all movies

#### Mean normalization

We will take all the movies and compute the mean for all the movies

Then we will take all these mean and subtract the mean from the original values

Then these values will become the new values for the algorithm and based on it the model will learn the parameters

And when predicting we will add the mean again

Now even if the algorithm learn the parameter to b 0 and 0 for the new user but by adding the mean they model will predict the rating

![](Images/mean_normalization_example.png)

By using the mean normalization the optimization algorithm also tend to work better
