## Reinforcement Learning

![](Images/reinforcement_learning.png)

Its not applied to lot of commercial products but its one of the pillars of ml

If we want to move for example an helicopter with the help of the reinforcement learning we will have the position of the helicopter and then based on the position we will decide how to move the control sticks to move the helicopter 

The position of the helicopter is called the state in the reinforcement learning and moving the control sticks is called action

For most of the real world examples like flying a helicopter or operating a robot the supervised learning approach does not work well as the conditions in the real world are not ideal and we can not have the data for all the conditions to train the model on

In the reinforcement learning we use a reward function which gives a positive feedback to the algorithm whenever it does a good work

Reinforcement learning is powerfull beacuse in it we tell the model what to do not how to do it and we tell it a reward function and it does that

## Return in the reinforcement Learning

![](Images/return.png)

The return captures that small rewards that you can get quickly can be better then the large reward that will take you a long time to get

Return is the sum of all the rewards from the given state to the terminal state plus the discount factor which is the number less then the 1 like 0.9

gamma is used to represent the discoutn factor

The gamma makes the algorithm impatient as the intial rewards that the algorithm gets are the 100% but as the step increase we take the cube of the gamma and so one so it will increase and then it will decrease the reward thus making the algorithm try to get the reward as soon as possible

Genrally the discount factor is close to the 1

![](Images/example_return.png)

We can see that based on that that always going to the left has better return then always going to the right but we do not have the always go left or right we can go left or right as well sometimes meaning right right left or left right left

![](Images/left_right.png)

See this is better then both as we are not only going in the one direction only

If the state has the negative reward then what the function does is that it delays the negative values and make them postpone in the future as much as possible

## Random environment

In some enviorments if we take a action into a new state its not compltely reliable for example window blowing etc

![](Images/example_stochastic.png)

Based on the given example see if we command robot to go the left there is a 90% chance it does it but 10% chance due to wind or any other issue it can fail it and go the opposite direction this is called the stochastic enviorment

So in the stochastic enviorment we do not focus on maximizing the return beacuse in the end its a random number we can get lucky or not lucky and instead we want to maximize the average of the return

So the job of the algorithm will be to choose the policy to maximize the average return
