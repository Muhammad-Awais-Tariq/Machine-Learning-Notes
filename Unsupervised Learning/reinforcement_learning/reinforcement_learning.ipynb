## Reinforcement Learning

![](Images/reinforcement_learning.png)

Its not applied to lot of commercial products but its one of the pillars of ml

If we want to move for example an helicopter with the help of the reinforcement learning we will have the position of the helicopter and then based on the position we will decide how to move the control sticks to move the helicopter 

The position of the helicopter is called the state in the reinforcement learning and moving the control sticks is called action

For most of the real world examples like flying a helicopter or operating a robot the supervised learning approach does not work well as the conditions in the real world are not ideal and we can not have the data for all the conditions to train the model on

In the reinforcement learning we use a reward function which gives a positive feedback to the algorithm whenever it does a good work

Reinforcement learning is powerfull beacuse in it we tell the model what to do not how to do it and we tell it a reward function and it does that

## Application of reinforcment learning

- Controlling the robots
- Factory optimization
- Stock trading
- Playing games
