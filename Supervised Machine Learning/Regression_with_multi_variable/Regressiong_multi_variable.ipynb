## Techniques for Gradient descent

### Feature scaling

If the feature has a large range and can choose the large values like from 300 to 2000 then the general rule of thumb is that the parameter value will be very small and if the feature has a relatively small range then the parameter value will b large

![image](Images/feature_and_parameter.png)

Now in this example we can see that the range of the size of house is large and the number of bedrooms is small now if we take the large parameter value for the size and small for the number of bedrooms we can see that the model predicts a value which is much larger then the actual value 

But if we take the small paramter for the size and large for the number we can see that the model prediction is more true


So this is the general case the model will generally pick the values that are small for the parameter with the large range and will pick the large parameter for the feature with small range

![image](Images/change_in_feature_and_parameter.png)

By the model we can see that the very small change in the w1 can have a large impact on the final value and it take a very large change in the w2 to cause a small change in the final value

### feature size and gradient descent

If we run the gradient descent in this we since the contour plot is thin and large the thing might happen that the gradient descent will have to run for a long time and it will go back and forth before it can finally reach the local minima

So to fix this its best to scale these values so they match and by scaling we mean that we do it such a way that both values take on a similar range

Now instead of having the range of size from 300 to 2000 and house from 1 to 5 we have scaled in such a way that both have same range and now we can see that the plot is also circular this way the gradient descent can easily find the minima thus increasing the efficency

## Implementing feature scaling

#### Deviding by maximum

If we have the values of different ranges an easy way to is to devide each value in the range by the maximum this will give us the value in the minimum range it mean devinding by the first element till the last so we will have the range between these values

![image](Images/Devding_by_max.png)

Now if we see the first one the range is from 0.15 to 1 beacuse 300 / 2000 is 0.15 and 2000/2000 is 1 so it take on this range of the values same is for the 0 and 5 so we have reduced the range of the values

#### Mean normalization

In the mean normalization we normalize the values in such a way that its center around the 0 and it takes on the values from -1 to 1 in both axis

![image](Images/Mean_normalization.png)

See in it we take the mean of the both parameters and then for each one them we subtract the mean from the value and then devide by the max - min and this way we will have the values with the points in the center

#### z score normalization

In the z score normalization we use the standar deviation we find the standard deviation and then using it we find the new range for each of these values

![images](Images/z_score_normalization.png)
