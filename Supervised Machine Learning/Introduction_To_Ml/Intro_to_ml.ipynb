## Mathematical formula of the gradient descent

Mathematically the gradient descent can be represented as in each step of the gradient descent the each value of the w will become 

$$w = w - \alpha \frac{\partial}{\partial w} J(w,b)$$

$$\alpha$$ is the symbol and is called the learning rate and it means the step how long or how short step we are going to take if the $$\alpha$$ is very large we are going to take very large step downhill and when the $$\alpha$$ is very small we are taking small steps to the decent

$$\frac{\partial}{\partial w} J(w,b)$$

This is the derivative and its telling as the direction meaning where we want to go

For the b we will write this equation as 

$$b = b - \alpha \frac{\partial}{\partial b} J(w,b)$$

Instead of taking the derivative of the w we take the derivative with b

We will continue updating this w and b values until we reach convergence its a point where further steps does not change mujhe values this is called the point on convergence

In the gradient descent we simutaneously update the both values meaning we update both at the same time

instead of directly assigning the values of the left side to the w and b we calculate them and store then in the temp and then assign that temp directly to both instead of computing
