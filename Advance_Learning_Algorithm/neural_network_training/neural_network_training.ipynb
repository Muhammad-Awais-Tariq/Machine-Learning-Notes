### Choosing the activation function

In the neural network we can chose the different functions for each of the neuron 

##### Choosing the function for output layer

Choosing the activation function for the neuron in the output layer is quite easy dependin on which kind of the output are we looking for if its 0 or 1 we can chose the sigmoid if only positive we can chose ReLU or if there is a range of values we can chose linear

![image](Images/activation_function_for_output.png)

##### Choosing the function for hidden layer

The most common choice for the hidden layers is the ReLU

##### Reasons for choosing ReLU

1. ReLu is faster to compute compare to the sigmoid
2. ReLU function is flat in only one place when the z is less then 0 but the sigmoid goes flat at 0 and 1 as well meaning on both places if we are using the gradient descent and a function is flat in only one place the gradient descent will perform much faster compare to if the function is flat at many places
