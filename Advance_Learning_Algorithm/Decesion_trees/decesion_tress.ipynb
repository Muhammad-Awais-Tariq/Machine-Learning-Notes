### Decision tree learning

When desiging the decision tree we have to decide
1. What feature to use at root node
2. We will split the group based on the values of the feature at root
3. The next step is focusing on the left part and choose the feature again to devide the group
4. we will split again based on the values
5. When we notice that all the group are cat and on the other are not cat we will create a leaf node that its cat and the other not cat
6. Repeat same for the other branch

![](Images/decesion_tree_learning.png)

## Choosing the feature to split

The split on the decesion tree will depend on the features that will reduce the entropy the most or the impurity the most this is called information gain as well

![](Images/weighted_average_entropy.png)

In order to find the feature to split we see the different features that are avaliable and then find the entropy for each of the branch and then we combine the entropy of the both branches and we combine the both entropy by taking the average of both

![](Images/conventional_way_of_split.png)

The conventional way is to compute the reduction in the entropy if we havnt used the spilt at all so instead of computing this average entropy what we do is to compute the entropy at each node and then compare it to previous one to find the reduction

The number computed after subtracting the intial entropy from the average is called the information gain

We compute this reduction beacuse one of the criteria to stop the further split is that if the reduction in the entropy is very small thats why finding the reduction in the entropy is important
