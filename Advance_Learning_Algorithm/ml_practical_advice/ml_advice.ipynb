## Evaluating the model

For example we have trained a model to predict the price of the house based on the size x so the model is f(x) and the model is polynomial

If we have only one feature like in this example the size and we are doing the prediction we can just plot the prediction and see that if the graphs is vigly that its not a good prediction

2. Using multiple feature to predict

For example if we are using x1 x2 x3 and x4 to predict the price of the house then it will be harder to plot the graph to see the desired outcome

![image](Images/multi_Feature.png)

So the first thing that we do is we split the data into two set the trainning set and the test set where the most of the data is in the training set

![image](Images/cost_test_for_Data.png)

So the next step us that we fit the parameters on the trainning set to minimize the cost function and then we compute the test error only on the training set and the training test error does not have the regulization term

And with the test error we also compute the training error as well 

The error calculation does not contain a regulization term beacuse in it we are not fitting to reduce the parameters we use regulization to fit the parameters but the error does not need parameter fitting

![image](Images/train_eval.png)

Now using the cost we can see for the multiple values as well that if the error in the trainining is very low but the error in the test is high it tell us that the model is fitted to the data and is not generalizing well
