### Diagonising bias and variance

If we underfit the data this is called that the data has high bias but if we data full accuratly its called high variance

![](Images/bias_varaince.png)

For the single feature we can plot to see that if the model has high bias or high variance but for the multiple features we can not do it

So the practical way for this is to look at the performance of the model on the training set as well as for the cross validation set

#### High Bias

If we train the model and the cost of the training sset of j of train is high and the cost of cross validation will also be high

We can find id that if the algorithm is not doing great on the training set and has high cost in the training set its the good indication that it has high bias

#### High varaince

If we train the model and the cost on the training set is low but when we test it on the cross validation set its cost is high it tell us that the model has high variance

We can identify it if the cost of cross validation is much higher then the cost of the training set

#### Middle ground

If the model is training and the cost on the training set is very low and the cost of the cross validation test is also very low then this tells us that the model just fits the data right
