The optimal way of selecting the model is instead of spiliting the data into two training and testing we split the data into three set one is training one is validation test and the last one is testing set

![image](Images/train_test_cross_validate.png)

![image](Images/dev_error.png)

Based on these three errors we can go on for model selection based on only these 

Now instead of evaluating directly on the train set we evaluate the models on the validation test and then we select the one with the least cost and then finally test it on the test set

![image](Images/general_methood.png)

Now this is the best methood for selecting the specific model and give the accurate infomation of how well it will perform on the new data as the test cost in this case has never been seen by the model it will give the accurate estimate of what the error will be
